# **OBJETIVO ANALÍTICO E HIPÓTESIS**

## **1 - Objetivo Analítico**
El propósito principal de este proyecto es analizar patrones climáticos globales mediante un dataset que incluye más de 75.000 registros y 41 variables relacionadas con condiciones atmosféricas, calidad del aire y factores astronómicos. Buscamos identificar relaciones significativas entre temperatura, humedad, velocidad del viento y contaminantes del aire (como PM2.5), con el fin de sentar las bases para futuros modelos predictivos que permitan anticipar eventos críticos, como altos niveles de contaminación o condiciones meteorológicas extremas.
Este análisis tiene como objetivo explorar el dataset para:

- Identificar los países con mayor nivel promedio de PM2.5 (polución).
- Investigar la posible relación entre la velocidad del viento y los niveles de PM2.5.
- Observar la distribución global de temperaturas.
- Determinar qué países presentan índices UV peligrosamente altos.
- Explorar la posible correlación entre humedad y niveles de PM2.5.

## **2 - Contexto Global**
 El cambio climático y el incremento de eventos climáticos adversos hacen cada vez más necesario comprender las variables ambientales que afectan la salud pública y el medio ambiente. Este proyecto toma datos recolectados a nivel mundial, permitiendo un análisis comparativo entre países y regiones para observar tendencias y posibles focos de riesgo.

## **3 - Problema a Abordar**
 Determinar qué países tienen valores promedio más elevados o variables climáticas críticas (temperatura, humedad, índice UV, PM2.5, etc.), identificando patrones que puedan requerir monitoreo o acciones preventivas para proteger a la población y al ambiente.

## **4 - Hipótesis**
- Concentración de contaminantes: Los países con mayores niveles promedio de PM2.5 estarán asociados a regiones con alta densidad poblacional, actividad industrial intensa o condiciones atmosféricas que favorecen la acumulación de contaminantes.

- Influencia de la latitud en el índice UV: El índice UV promedio será significativamente más alto en países ubicados en latitudes cercanas al ecuador debido a la mayor exposición directa a la radiación solar.

- Relación entre variables climáticas y PM2.5: La velocidad del viento y la humedad presentarán correlaciones significativas con los niveles de PM2.5, afectando su dispersión y concentración.

- Distribución térmica: La mayoría de los registros de temperatura se concentrarán en rangos asociados a climas templados y tropicales, reflejando la diversidad geográfica de los datos.

## **Filtrado de Outliers**

```python
# ---------- Antes del filtrado ----------
plt.figure(figsize=(12,5))

# Boxplot Velocidad del viento (antes)
plt.subplot(1,2,1)
sns.boxplot(x=df["wind_kph"], color='skyblue')
plt.axvline(150, color='red', linestyle='--', linewidth=2, label='Umbral 150 kph')
plt.title("Velocidad del viento (antes del filtrado)", fontsize=14, fontweight='bold')
plt.legend()

# Boxplot PM2.5 (antes)
plt.subplot(1,2,2)
sns.boxplot(x=df["air_quality_PM2.5"], color='lightgreen')
plt.axvline(500, color='red', linestyle='--', linewidth=2, label='Umbral 500 µg/m³')
plt.title("PM2.5 (antes del filtrado)", fontsize=14, fontweight='bold')
plt.legend()

plt.tight_layout()
plt.show()
```

Tras analizar el dataset con `describe().T` y algunos gráficos iniciales,
se detectaron valores extremos en variables clave como:

- `wind_kph`: 75% de los datos están debajo de 18.4, pero hay registros de hasta 2963 kph.
- `air_quality_PM2.5`: 75% de los datos están debajo de 33, pero hay registros superiores a 4000.

Estos valores pueden corresponder a errores de sensores o registros atípicos
que distorsionan el análisis.

Por lo tanto, se aplicaron los siguientes filtros para remover outliers:

- **Velocidad del viento (`wind_kph`) menor a 150 kph**
- **Concentración de PM2.5 menor a 500 µg/m3**

Esto permitió mantener los registros dentro de un rango físicamente plausible.


```python
df_filt = df[(df["wind_kph"] < 150) & (df["air_quality_PM2.5"] < 500)]
df_filt2 = df[df["air_quality_PM2.5"] < 500]
print("Registros antes del filtro:", len(df))
print("Registros después del filtro:", len(df_filt))
```

```python
# ---------- Despues del filtrado ----------
plt.figure(figsize=(12,5))

# Boxplot Velocidad del viento (después)
plt.subplot(1,2,1)
sns.boxplot(x=df_filt["wind_kph"], color='skyblue')
plt.axvline(150, color='red', linestyle='--', linewidth=2, label='Umbral 150 kph')
plt.xlim(0, 200)  # Ajustar rango para ver mejor la caja
plt.title("Velocidad del viento (después del filtrado)", fontsize=14, fontweight='bold')
plt.legend()

# Boxplot PM2.5 (después)
plt.subplot(1,2,2)
sns.boxplot(x=df_filt["air_quality_PM2.5"], color='lightgreen')
plt.axvline(500, color='red', linestyle='--', linewidth=2, label='Umbral 500 µg/m³')
plt.xlim(0, 600)  # Ajustar rango
plt.title("PM2.5 (después del filtrado)", fontsize=14, fontweight='bold')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Países con mayor PM2.5 promedio
pm25_prom = df.groupby("country")["air_quality_PM2.5"].mean().sort_values(ascending=False)
pm25_prom.head(10)
```

```python
# Gráfico 2: Top 10 Países con mayor PM2.5 promedio, gráfico de Área (Area Chart).
pm25_top10 = pm25_prom.head(10)

plt.figure(figsize=(10,5))
plt.plot(pm25_top10.index, pm25_top10.values, color='purple', marker='o')
plt.fill_between(pm25_top10.index, pm25_top10.values, color='purple', alpha=0.3)

plt.title("Top 10 Países con mayor PM2.5 promedio", fontsize=14, fontweight='bold')
plt.ylabel("PM2.5 promedio")
plt.xlabel("País")
plt.xticks(rotation=45)

# Etiquetas con los valores
for x, y in zip(pm25_top10.index, pm25_top10.values):
    plt.text(x, y+2, f"{y:.1f}", ha='center')

plt.tight_layout()
plt.show()
```

**Insight:** Este gráfico muestra que Chile lidera el ranking de países con mayor concentración promedio de PM2.5, seguido por China, Arabia Saudita e India. Estos elevados niveles pueden estar relacionados con actividades industriales, urbanización acelerada o condiciones geográficas que favorecen la acumulación de contaminantes. Este hallazgo resalta la necesidad de investigar políticas ambientales y factores locales que contribuyen a estos valores, ya que el PM2.5 está directamente vinculado a problemas respiratorios y cardiovasculares.
